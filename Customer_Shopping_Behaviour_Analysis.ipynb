For checking missing values/null values

```python
df.isnull().sum()
# ab null values ko bharne ka tin tareeka hai 1. kuch bhi daal do 
# 2. mean daalo->ye outliers se effect hota hai 
# 3. median daalo->ye outliers se effect nhi hota hai 
# lekin sbme overall medain leke nhi fill kr skte dataset biased ho jayega 
# to har category me median review rating of each item lo and replace kro ushko accordingly
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

# data analyse kro and then put kro usme else data will be biased
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
# kuch ko rename kr dete hai 
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

#ab column name wagaira shi hogya hai ab feature engneering krne ki baari hai

```python
# create a new column to group customers into age group: 1. young adult, adult , middle-aged,senior
labels=['Young Adult','Adult','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

# yaha kya hua hai basically qcut me pure data segment ko 4 euqal parts me break kr diya hai aisa ki har part me same no. of ppl rhe 
# and ek hota hai cut function wo equal interval me divide krta hai interval/ range yaha no. of ppl in one group matter nhi krta .

```python
# create a column with purchase frequency days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
